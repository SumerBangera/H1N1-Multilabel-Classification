```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

from preprocessor import Preprocessor
pp = Preprocessor()
```

```python
training_set_features = pd.read_csv("training_set_features.csv", index_col = "respondent_id")
training_set_labels = pd.read_csv("training_set_labels.csv", index_col="respondent_id")
test_set_features = pd.read_csv("test_set_features.csv", index_col = "respondent_id")
```

```python
pp.data_prep_check(df = training_set_features)
```

```python
train_df = pp.preprocess_data(training_set_features)
train_labels = pd.read_csv("training_set_labels.csv", index_col="respondent_id")
test_df = pp.preprocess_data(test_set_features)
```

```python
# train_df.to_pickle("train_df.pkl")
# test_df.to_pickle("test_df.pkl")
# train_labels.to_pickle("train_labels.pkl")

train_df = pd.read_pickle("train_df.pkl")
test_df = pd.read_pickle("test_df.pkl")
train_labels = pd.read_pickle("train_labels.pkl")
```

```python
pp.data_prep_check(train_df)
```

### Modeling and Analysis

```python
from modelling import Modelling
from models import Models

from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier

import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 6  

mo = Models()
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_df, train_labels, 
                                                  test_size = 0.2, 
                                                  shuffle = True,    
                                                 # stratify=train_labels,   #without stratifying the ratios are still maintained
                                                  random_state=RANDOM_SEED)
```

Logistic Regression

```python
logit_model = mo.logit_model(X_train, y_train, X_val,y_val)
```

Decision Tree

```python
dt_model = mo.dt_model(X_train, y_train, X_val,y_val)
```

Random Forest

```python
rf_model = mo.rf_model(X_train, y_train, X_val,y_val)
```

Extra Trees Classifier

```python
etc_model = mo.etc_model(X_train, y_train, X_val,y_val)
```

Light GBM

```python
lgb_model = mo.lgb_model(X_train, y_train, X_val,y_val)
```

Catboost

```python
catboost_model = mo.catboost_model(X_train, y_train, X_val,y_val)
```

Extra Trees Classifier with RanomizedSearchCV

```python
# Using RandomizedSearchCV()
etc_model = mo.etc_model(X_train, y_train, X_val,y_val, cv_type = 'randsearch')
```

#### Save all models to disk

```python
mo.save_model(model = logit_model,name = "logit_model")
mo.save_model(model = dt_model, name = "dt_model")
mo.save_model(model = rf_model,name = "rf_model")
mo.save_model(model = etc_model, name = "etc_model")
mo.save_model(model = lgb_model,name = "lgb_model")
mo.save_model(model = catboost_model, name = "catboost_model")
```

#### Test file for submission

```python
test_probas = final_model.predict_proba(test_df)
test_probas
```
